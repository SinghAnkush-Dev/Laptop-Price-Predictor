Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# Load dataset
df = pd.read_csv("../data/Laptop_price.csv")

df.head()
```

Dataset Shape

```python
print(df.shape)
```

Dataset Info

```python
df.info()
```

Statistics

```python
df.describe()
```

Missing Values

```python
df.isnull().sum()
```

Brand Distribution

```python
df["Brand"].value_counts()
```

Price Distribution

```python
plt.figure(figsize=(8,5))
plt.hist(df["Price"], bins=20)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

Brand VS Price

```python
fig = px.box(
    df,
    x="Brand",
    y="Price",
    title="Brand vs Price"
)

fig.show()
```

RAM VS PRICE

```python
fig = px.scatter(
    df,
    x="RAM_Size",
    y="Price",
    color="Brand",
    title="RAM vs Price"
)

fig.show()
```

Storage vs Price

```python
fig = px.scatter(
    df,
    x="Storage_Capacity",
    y="Price",
    color="Brand",
    title="Storage vs Price"
)

fig.show()
```

Correlation Matrix

```python
corr = df.corr(numeric_only=True)
corr

```

Correlation Heatmap

```python
plt.figure(figsize=(8,6))

plt.imshow(corr, cmap="Blues")

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)

plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")

plt.show()
```
